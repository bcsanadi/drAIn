```python
import matplotlib.pyplot as plt

# Define the actions and their equivalent AI prompts
actions = [
    "Took a shorter shower",
    "Turned off water\nwhile brushing teeth or\nwashing hands/face",
    "Ran a full load in the\ndishwasher or\nwashing machine",
    "Fixed a dripping faucet",
    "Scraped dishes\ninstead of rinsing\nbefore loading\ndishwasher"
]

prompts_refilled = [10, 3, 6, 1_930, 17]

# Create the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(actions, prompts_refilled, color="#cbac86", edgecolor="black")

# Add title and labels
plt.title("AI Prompts 'Refilled' by Everyday Water-Saving Actions", fontsize=14, fontweight="bold")
plt.xlabel("Action", fontsize=12)
plt.ylabel("Equivalent 100-Word AI Prompts Refilled", fontsize=12)
plt.xticks(rotation=15, ha="right")

# Annotate each bar with its value
for bar, value in zip(bars, prompts_refilled):
    height = bar.get_height()
    offset = max(height * 0.01, 1)  # 1% of bar height or at least 1 unit
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + offset,
        f"{value:,}",
        ha="center",
        va="bottom",
        fontsize=10,
        fontweight="bold"
    )

# Optional: use log scale to handle the large difference in values
plt.yscale("log")
plt.text(3.6, 1.2, "", fontsize=9, style="italic", color="gray")

plt.tight_layout()
plt.show()
```
